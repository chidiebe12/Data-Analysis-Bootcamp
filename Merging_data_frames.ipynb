```python
#MERGING DATAFRAMES
```

```python
import pandas as pd
```

```python
data_1={
    "Employee_id":[1,2,3,4],
    "Names":["Fortune","Samuel","Paul","Martins"],
    "Dept_id":[101,102,103,104]
}
df_1 = pd.DataFrame(data_1)
df_1
```

```python
data_2={
    "Dept_names":["Sales","Marketing","HR","IT"],
    "Names":["Fortune","Samuel","David","Daniel"],
    "Dept_id":[101,102,106,107]
}
df_2 = pd.DataFrame(data_2)
df_2
```

```python
df_1.merge(df_2)
```

```python
df_1.merge(df_2,how="inner")
```

```python
df_1.merge(df_2,how="inner",on="Names")
```

```python
df_1.merge(df_2,how="outer")
```

```python
df_1.merge(df_2,how="left")
```

```python
df_1.merge(df_2,how="right")
```

```python
df_1.merge(df_2,how="cross")
```

```python
df_1.join(df_2,how="inner",lsuffix="_left",rsuffix="_right")
```

```python
df_1.set_index("Names").join(df_2.set_index("Names"),on="Names",how="inner",lsuffix="_left",rsuffix="_right")
```

```python
df_1.set_index("Names").join(df_2.set_index("Names"),on="Names",how="outer",lsuffix="_left",rsuffix="_right")
```

```python
df_1.set_index("Names").join(df_2.set_index("Names"),on="Names",how="right",lsuffix="_left",rsuffix="_right")
```

```python
df_1.set_index("Names").join(df_2.set_index("Names"),on="Names",how="left",lsuffix="_left",rsuffix="_right")
```

```python
pd.concat([df_1,df_2])
```

```python
pd.concat([df_1,df_2],axis=1)
```

```python
pd.concat([df_1,df_2],axis=0)
```
